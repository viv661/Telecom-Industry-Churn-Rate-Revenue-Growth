```python
df = pd.read_csv('telecom_churn.csv',usecols=['telecom_partner','gender','age','state','num_dependents','estimated_salary','calls_made','sms_sent','data_used','churn'])
```

```python
df.head()
```

```python
df.describe()
```

```python
def categorize_age(age):
    if 18 <= age <= 32:
        return 'A'
    elif 33 <= age <= 46:
        return 'B'
    elif 47 <= age <= 60:
        return 'C'
    elif 61 <= age <= 74:
        return 'D'
    else:
        return None

df['age_groups'] = df['age'].apply(categorize_age)
```

```python
def categorize_salary(salary):
    if  20000<= salary <= 52585:
        return 'Basic'
    elif 52585 < salary <= 84990:
        return 'Plus'
    elif 84990 < salary <= 117488:
        return 'Prime'
    elif 117488 < salary <= 149999:
        return 'Elite'
    else:
        return None  

df['estimated_salary_club'] = df['estimated_salary'].apply(categorize_salary)
```

```python
df.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(14, 6))
sns.countplot(x='state', hue='telecom_partner', data=df, palette='Set2')
plt.xticks(rotation=90)
plt.title('Telecom Partner Distribution Across States')
plt.xlabel('State')
plt.ylabel('Customer Count')
plt.legend(title='Telecom Partner')
plt.show()
```

```python
percentage_table = pd.crosstab(df['state'], df['telecom_partner'], normalize='index') * 100
plt.figure(figsize=(12, 8))
sns.heatmap(percentage_table, cmap='YlOrBr', annot=True, fmt=".1f")
plt.title('Percentage Distribution of Telecom Partners by State')
plt.xlabel('Telecom Partner')
plt.ylabel('State')
plt.show()
```

```python
partner_counts = df['telecom_partner'].value_counts()
plt.pie(partner_counts, labels=partner_counts.index, autopct='%1.1f%%', startangle=90, colors=['orange','yellow','skyblue','green'])
plt.title('Telecom Partner Distribution')
plt.show()
```

```python
plt.figure(figsize=(14, 6))
sns.countplot(x='age_groups', hue='telecom_partner', data=df, palette='Set2')
plt.xticks(rotation=90)
plt.title('Telecom Partner Distribution Across age_groups')
plt.xlabel('age_groups')
plt.ylabel('Customer Count')
plt.legend(title='Telecom Partner')
plt.show()
```

```python
plt.figure(figsize=(14, 6))
sns.countplot(x='estimated_salary_club', hue='telecom_partner', data=df, palette='Set2')
plt.xticks(rotation=90)
plt.title('Telecom Partner Distribution Across estimated_salary_club')
plt.xlabel('estimated_salary_club')
plt.ylabel('Count')
plt.legend(title='Telecom Partner')
plt.show()
```

```python
sns.countplot(x='age_groups', hue='churn', data=df, palette='coolwarm')
plt.title('Churn Rate by Age Group')
plt.xlabel('Age Groups')
plt.ylabel('Count')
plt.show()
```

```python
sns.countplot(x='estimated_salary_club', hue='churn', data=df, palette='viridis')
plt.title('Churn vs Salary Club')
plt.xlabel('Salary Club')
plt.ylabel('Count')
plt.show()
```

```python
sns.set(style="whitegrid", palette="muted")
plt.figure(figsize=(4,3))
sns.barplot(data=df, x='telecom_partner', y='churn', estimator=lambda x: sum(x)/len(x)*100)
plt.ylabel("Churn Rate (%)")
plt.title("Churn Rate by Telecom Partner")
plt.show()
```

```python
plt.figure(figsize=(7,5))
sns.barplot(data=df, x='age_groups', y='churn', estimator=lambda x: sum(x)/len(x)*100)
plt.ylabel("Churn Rate (%)")
plt.title("Churn Rate by Age Group")
plt.show()
```

```python
revenue_share = df.groupby('estimated_salary_club')['estimated_salary'].sum()
plt.figure(figsize=(6,6))
plt.pie(revenue_share, labels=revenue_share.index, autopct='%1.1f%%', startangle=140)
plt.title("Revenue Contribution by Salary Club")
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(data=df, x='estimated_salary', y='data_used', hue='churn', size='calls_made', sizes=(50,200))
plt.title("High-value Churners: Salary vs Data Usage")
plt.show()
```

```python
fig, axes = plt.subplots(1, 2, figsize=(14,5))
sns.boxplot(data=df, x='churn', y='calls_made', ax=axes[0])
axes[0].set_title("Calls Made vs Churn")
sns.boxplot(data=df, x='churn', y='sms_sent', ax=axes[1])
axes[1].set_title("SMS Sent vs Churn")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(data=df, x='data_used', hue='churn', bins=20, kde=True, palette="coolwarm")
plt.title("Data Usage Distribution by Churn")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(data=df, x='estimated_salary', hue='churn', bins=20, kde=True, palette="coolwarm")
plt.title("Estimated Salary Distribution by Churn")
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.heatmap(df[['age','num_dependents','estimated_salary','calls_made','sms_sent','data_used','churn']].corr(),
            annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```
