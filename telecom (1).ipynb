```python
df = pd.read_csv('telecom_churn.csv',usecols=['telecom_partner','gender','age','state','num_dependents','estimated_salary','calls_made','sms_sent','data_used','churn'])
```

```python
df
```

```python
df.describe()
```

```python
def categorize_age(age):
    if 18 <= age <= 32:
        return 'A'
    elif 33 <= age <= 46:
        return 'B'
    elif 47 <= age <= 60:
        return 'C'
    elif 61 <= age <= 74:
        return 'D'
    else:
        return None

df['age_groups'] = df['age'].apply(categorize_age)
```

```python
def categorize_salary(salary):
    if  20000<= salary <= 52585:
        return 'Basic'
    elif 52585 < salary <= 84990:
        return 'Plus'
    elif 84990 < salary <= 117488:
        return 'Prime'
    elif 117488 < salary <= 149999:
        return 'Elite'
    else:
        return None  

df['estimated_salary_club'] = df['estimated_salary'].apply(categorize_salary)
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
data = df.copy()
```

```python
label_encoders = {}
for col in data.select_dtypes(include=['object']).columns:
    le = LabelEncoder()
    data[col] = le.fit_transform(data[col])
    label_encoders[col] = le 
```

```python
X = data.drop(columns=['churn'])
y = data['churn']
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42, stratify=y
)
```

```python
model = LogisticRegression(max_iter=500)
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
print("Accuracy:", accuracy_score(y_test, y_pred))
```

```python
total_customers = len(df)
churned_customers = df['churn'].sum()
current_churn_rate = churned_customers / total_customers * 100
print(f"Current Churn Rate: {current_churn_rate:.2f}%")
print(churned_customers)
```

```python
predicted_churners = churned_customers * accuracy_score(y_test, y_pred)
print(predicted_churners)
```

```python
intervention_success = 0.5
saved_customers = predicted_churners * intervention_success
```

```python
new_churners = churned_customers - saved_customers
new_churn_rate = new_churners / total_customers
print(f"New churn rate: {new_churn_rate:.2%}")
```
